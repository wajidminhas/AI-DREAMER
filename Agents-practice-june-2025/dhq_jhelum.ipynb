```python
from agents import AsyncOpenAI, OpenAIChatCompletionsModel, Agent, enable_verbose_stdout_logging, Runner, function_tool

enable_verbose_stdout_logging()

# Our existing doctors_dict, representing doctors on duty/available

@function_tool
def display_on_duty_doctors() -> str:
  dhq_jehlum = {
    'Dr. Saira Khan': 'MBBS',
    'Dr. Shama Raja': 'MBBS',
    'Dr. Farah Bashir': 'MBBS',
    'Dr. Saeeda Itrat Batool': 'MBBS',
    'Dr. Beenish Karamat': 'MBBS',
    'Dr. Bisma Fater': 'MBBS',
    'Dr. Misbah Ul Ain': 'MBBS',
    'Dr. Nadia Noureen': 'MBBS',
    'Dr. Mehreen Khalid': 'MBBS',
    'Dr. Kiran Rani': 'MBBS',
    'Dr. Javeria Azhar': 'MBBS',
    'Dr. Anam Jamil': 'MBBS'
  }
  for doctor, designation in dhq_jehlum.items():
      return (f"{doctor} - {designation}")







@function_tool
def get_doctor(name : str, designation : str) -> str:
  """
  get the best doctor for the patient with the given name and designation

  arg:
    name: str
    designation: str

  return:
    str
  """

external_client = AsyncOpenAI(
    api_key = gemini_api_key,
    base_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
)

model = OpenAIChatCompletionsModel(
    openai_client = external_client,
    model = "gemini-2.0-flash"
)

triage_agent = Agent(
    name = "Triage Agent",
    instructions= "you are a helpful assistent to solve about doctor and facilities available",
    model = model,
    tools = [get_doctor, display_on_duty_doctors]

)

result = Runner.run_sync(
    triage_agent,
    "is Dr. Javeria Azhar mbbs or Dr. Kiran Rani on duty ?"
)

print(result)
```
