## LLM With Tools Csll

```python
from agents import AsyncOpenAI, OpenAIChatCompletionsModel, Agent, enable_verbose_stdout_logging, Runner, function_tool, ModelSettings

enable_verbose_stdout_logging()

# Our existing doctors_dict, representing doctors on duty/available

doctors_dict = {
    'Dr. Saira Khan': 'MBBS',
    'Dr. Shama Raja': 'MBBS',
    'Dr. Farah Bashir': 'MBBS',
    'Dr. Saeeda Itrat Batool': 'MBBS',
    'Dr. Beenish Karamat': 'MBBS',
    'Dr. Bisma Fater': 'MBBS',
    'Dr. Misbah Ul Ain': 'MBBS',
    'Dr. Nadia Noureen': 'MBBS',
    'Dr. Mehreen Khalid': 'MBBS',
    'Dr. Kiran Rani': 'MBBS',
    'Dr. Javeria Azhar': 'MBBS',
    'Dr. Anam Jamil': 'MBBS'
  }

@function_tool(is_enabled=False)
def display_on_duty_doctors():
    """
    Provides a comprehensive list of all doctors currently on duty at the hospital,
    including their degrees. This function is useful when the user wants to
    see ALL available doctors without specifying a name.

    This function does not require any parameters.
    """
    # Access the doctors_dict directly
    hospital_doctors_data = doctors_dict

    output_lines = ["--- Doctors Currently On Duty ---"]

    if not hospital_doctors_data:
        output_lines.append("No doctors are currently on duty according to our records.")
    else:
        for doctor_name in hospital_doctors_data:
            degree = hospital_doctors_data[doctor_name]
            output_lines.append(f"- {doctor_name} ({degree})")
        output_lines.append("\nPlease choose from the list above for your consultation.")

    # Return the output as a single string, which the LLM can then present
    return "\n".join(output_lines)

@function_tool
def get_doctor():
  """
  get the best doctor for the patient with the given name and designation

  arg:
    name: str
    designation: str

  return:
    str
  """
  name = input("Enter the name of the doctor: ")
  designation = input("Enter the designation of the doctor: ")
  return f"The best doctor for {name} with {designation} is Dr. Saira Khan"

external_client = AsyncOpenAI(
    api_key = gemini_api_key,
    base_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
)

model = OpenAIChatCompletionsModel(
    openai_client = external_client,
    model = "gemini-2.0-flash"
)

triage_agent = Agent(
    name = "Triage Agent",
    instructions= "you are a helpful assistent to solve about doctor and facilities available, my name is wajid shabbir so book my appointment",
    model = model,
    tools = [get_doctor, display_on_duty_doctors],
    # model_settings=ModelSettings(tool_choice="required"),
    # tool_use_behavior="stop_on_first_tool"

)

result = Runner.run_sync(
    triage_agent,
    "who doctor is best in Medical specialist?",
    max_turns=1
)

print(f"\n[RUNNER RESULT]")
print(result)
print(f"\n[FINAL OUTPUT]\n")
print(result.final_output)
```
